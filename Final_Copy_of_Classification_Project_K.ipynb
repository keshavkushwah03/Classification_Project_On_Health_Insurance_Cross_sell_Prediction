### Hypothetical Statement - 1

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

**Null Hypothesis (H0):**The average age of customers who responded positively to the vehicle insurance offer is the same as the average age of customers who responded negatively.

**Alternate Hypothesis (H1):** The average age of customers who responded positively to the vehicle insurance offer is grater then from the average age of customers who responded negatively.

#### 2. Perform an appropriate statistical test.

````python
# Calculate the average age of customers who responded positively (Response = 1)
average_age_positive = int(copy_df[copy_df['Response'] == 1]['Age'].mean())

# Calculate the average age of customers who responded negatively (Response = 0)
average_age_negative = int(copy_df[copy_df['Response'] == 0]['Age'].mean())

print("Average Age of Customers who Responded Positively:", average_age_positive)
print("Average Age of Customers who Responded Negatively:", average_age_negative)
````

````python
# Perform Independent Samples t-test to obtain P-Value

from scipy.stats import ttest_ind
# Perform Independent Samples t-test
t_stat, p_value = ttest_ind(copy_df[copy_df['Response'] == 1]['Age'], copy_df[copy_df['Response'] == 0]['Age'], equal_var=False)

print(p_value)

if p_value > 0.05:     # significance level
    print('Accept Null Hypothesis')
else:
    print('Rejected Null Hypothesis')
````

### 9. Handling Imbalanced Dataset

##### Do you think the dataset is imbalanced? Explain Why.

The dataset is imbalanced because the count of class 0 is significantly higher than the count of class 1 (from chart 1). This can lead to biased model predictions and poorer performance for the minority class.

````python
# Handling Imbalanced Dataset

from sklearn.utils import resample

# Separate majority and minority classes
majority_class = copy_df[copy_df['Response'] == 0]
minority_class = copy_df[copy_df['Response'] == 1]

# Upsample the minority class
minority_upsampled = resample(minority_class, replace=True, n_samples=len(majority_class), random_state=0)

# Combine majority class with upsampled minority class
balanced_df = pd.concat([majority_class, minority_upsampled])

# Separate the target variable from the features
X_balanced = balanced_df.drop('Response', axis=1)
y_balanced = balanced_df['Response']

# Class distribution before balancing
print("Class distribution before balancing:")
print(copy_df['Response'].value_counts())

# Class distribution after balancing
print("\nClass distribution after balancing:")
print(balanced_df['Response'].value_counts())
````
